#Plot 그리기

1. PAE plot

```python
from pathlib import Path
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
scores = json.loads(Path("test_8dda0_3_scores_rank_005_alphafold2_multimer_v3_model_1_seed_000.json").read_text())
```

```python
from matplotlib import colormaps
list(colormaps)
```

```python
plt.imshow(np.asarray(scores["pae"]), cmap='PuBu_r')
plt.xlabel("Scored residue")
plt.ylabel("Aligned residue")
plt.colorbar()
plt.savefig('CAB39-OSR1_PAE.pdf', dpi=600)
plt.show()
```

2. pLDDT plot

```python
plt.plot(numpy.asarray(scores["plddt"]))
plt.xlabel("Residue")
plt.ylabel("average pLDDT")
plt.show()
```
